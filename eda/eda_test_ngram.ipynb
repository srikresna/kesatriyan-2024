```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/unlabeled_dataset_clean_1.csv')
df
```

```python
# analisa ngram text
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

# ngram_range=(1, 1) -> unigram
# ngram_range=(2, 2) -> bigram
# ngram_range=(3, 3) -> trigram
# ngram_range=(1, 3) -> unigram, bigram, trigram
unigram = TfidfVectorizer(ngram_range=(1, 1))
bigram = TfidfVectorizer(ngram_range=(2, 2))
trigram = TfidfVectorizer(ngram_range=(3, 3))
unigram_bigram = TfidfVectorizer(ngram_range=(1, 2))
unigram_bigram_trigram = TfidfVectorizer(ngram_range=(1, 3))
```

```python
# unigram
unigram_matrix = unigram.fit_transform(df['processed_text'])
unigram_matrix = unigram_matrix.toarray()
unigram_matrix = pd.DataFrame(unigram_matrix, columns=unigram.get_feature_names_out())

# visualisasi unigram
unigram_matrix.sum().sort_values(ascending=False).head(10).plot(kind='barh')
plt.title('Top 10 Unigram')
plt.show()
```

```python
# bigram
bigram_matrix = bigram.fit_transform(df['processed_text'])
bigram_matrix = bigram_matrix.toarray()
bigram_matrix = pd.DataFrame(bigram_matrix, columns=bigram.get_feature_names_out())

# visualisasi bigram
bigram_matrix.sum().sort_values(ascending=False).head(10).plot(kind='barh')
plt.title('Top 10 Bigram')
plt.show()
```

```python
# trigram
trigram_matrix = trigram.fit_transform(df['processed_text'])
trigram_matrix = trigram_matrix.toarray()
trigram_matrix = pd.DataFrame(trigram_matrix, columns=trigram.get_feature_names_out())

# visualisasi trigram
trigram_matrix.sum().sort_values(ascending=False).head(10).plot(kind='barh')
plt.title('Top 10 Trigram')
plt.show()
```

```python
# unigram bigram
unigram_bigram_matrix = unigram_bigram.fit_transform(df['processed_text'])
unigram_bigram_matrix = unigram_bigram_matrix.toarray()
unigram_bigram_matrix = pd.DataFrame(unigram_bigram_matrix, columns=unigram_bigram.get_feature_names_out())

# visualisasi unigram bigram
unigram_bigram_matrix.sum().sort_values(ascending=False).head(10).plot(kind='barh')
plt.title('Top 10 Unigram Bigram')
plt.show()
```

```python
# unigram bigram trigram
unigram_bigram_trigram_matrix = unigram_bigram_trigram.fit_transform(df['processed_text'])
unigram_bigram_trigram_matrix = unigram_bigram_trigram_matrix.toarray()
unigram_bigram_trigram_matrix = pd.DataFrame(unigram_bigram_trigram_matrix, columns=unigram_bigram_trigram.get_feature_names_out())

# visualisasi unigram bigram trigram
unigram_bigram_trigram_matrix.sum().sort_values(ascending=False).head(10).plot(kind='barh')
plt.title('Top 10 Unigram Bigram Trigram')
plt.show()
```
